```python
def check_input(element):
    return type(element) == str


def get_substring(s):
    max_sub = ""
    for i in range(len(s)):
        for j in range(i+1, len(s)+1):
            sub = s[i:j]
            unique_chars = ""
            repeat = False
            for c in sub:
                if c in unique_chars:
                    repeat = True
                    break
                else:
                    unique_chars += c
            if not repeat and len(sub) > len(max_sub):
                max_sub = sub
    return max_sub


def main(arr):
    max_len = -1
    max_index = -1
    max_string = ""


    for i in range(len(arr)):
        if check_input(arr[i]):
            if len(arr[i]) > max_len:
                max_len = len(arr[i])
                max_index = i
                max_string = arr[i]

    if max_index == -1:
        print(-1, "")
    else:
        result = get_substring(max_string)
        print(max_index, result)
```
